## Exercise 2
### b)
- Only consider items {1,2,...,20}
- Every pair of items with item 1 is frequent, as 1 appears in all the baskets
- Every pair of items with an even id and item 2 are frequent, as 2 appears always with even items
- Every pair of items with an item divisible by 3 and item 3 are frequent
- and so on...


Pattern 1: Every item i, with 1 <= i <= 20, forms a frequent pair with an item j, with j < i and i % j = 0.

Pattern 2: Every pair of items i, j, with i * j <= 20 is frequent, as 20 is frequent

General rule: In a group of items, if the common product is smaller than 20, the group is frequent.
